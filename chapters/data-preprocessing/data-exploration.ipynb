```python
categorical_features = ['Make', 'Model', 'Color', 'Body_Type', 'Drive_Type',
       'Transmission', 'Emission_Class', 'Fuel', 'Category',
       'Interior_Material', 'Interior_Color', 'Paint_Type',  'Non_Smoker_Car',
       'Other_Energy_Sources', 'Doors']
```

```python
date_features = ['Production_Date', 'Build_Year', 'Technical_Inspection', 'Last_Service']
```

```python
numerical_features = [feature for feature in df.columns if feature not in categorical_features + date_features]
numerical_features
```

```python
df['Make'].astype('category').cat.codes
```

```python
# Check the time
start_time = time.time()
for _ in range(20):
    df.groupby(['Make', 'Model', 'Color', 'Body_Type'])['Mileage'].mean()
time.time() - start_time
```

```python
# Requires more memory
df['Make'].nbytes
```

```python
for feature in categorical_features:
    df[feature] = df[feature].astype('category')
```

```python
# Faster (if observed = True)
start_time = time.time()
for _ in range(20):
    df.groupby(['Make', 'Model', 'Color', 'Body_Type'], observed=True)['Mileage'].mean()
time.time() - start_time
```

```python
# Slower (if observed = False)
start_time = time.time()
for _ in range(20):
    df.groupby(['Make', 'Model', 'Color', 'Body_Type'])['Mileage'].mean()
time.time() - start_time
```

```python
df_temp = df.copy()
for feature in date_features:
    df_temp[feature] = pd.to_datetime(df[feature], errors='coerce')
df_temp.isna().sum().sum()
```
